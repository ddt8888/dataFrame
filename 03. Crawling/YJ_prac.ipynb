```python
import requests
from bs4 import BeautifulSoup

url = 'http://www.yjjob.or.kr/p/?j=23'

res = requests.get(url)
soup = BeautifulSoup(res.content, 'html.parser')
```

```python
print(soup)
```

```python
list_genre = []
list_title = []
list_periodb = []
list_periodr = []
list_date = []
list_cost = []
list_money = []
```

```python
yj_title = soup.select('tbody > tr > td.lh_140.v_top.relative > a.p100 > span.font14.bold')
for i in yj_title:
    list_title.append('교육과정명 : ' + i.text)
```

```python
print(len(list_title))
for i in list_title:
    print(i)
```
