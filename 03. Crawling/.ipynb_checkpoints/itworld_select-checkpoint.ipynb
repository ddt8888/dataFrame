```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.itworld.co.kr/main/'

res = requests.get(url)
soup = BeautifulSoup(res.content, 'html.parser')
```

```python
print(soup)
```

```python
itw = soup.select('div')
print(itw)
```

```python
it = soup.select('ul > li > div')
print(len(it))
print(it)
```

```python
list_all = []
list_titles = []
list_time = []
```

```python
titles = soup.select('ul.default_font.color_deep_gray.hot_news_list_.title_h3 > \
                    li.font_bold > div.font_limit.fl')
for title in titles:
    list_titles.append(title.text.strip())
```

```python
print(titles)
```

```python
for title in titles:
    print(title.text)
```

```python
times = soup.select('ul.default_font.color_deep_gray.hot_news_list_.title_h3 > \
                    li.font_bold > div.news_list_time.font_11.color_black.default_small_font.fr')
for time in times:
    list_time.append(time.text.strip())
```

```python
for time in times:
    print(time.text)
```

```python
for i in range(len(list_titles)) :
    list_i.append('제목 : ' + str(list_titles[i]) + ', 게시시간 : ' + str(list_times[i]))
```
