```python
import matplotlib.pyplot as plt
import numpy as np
```

#### 범례(Lengend)

- 범례(Lengend)는 그래프에 데이터의 종류를 표시하기 위한 텍스트

## 기본 사용
- plot() 함수 안에 label=''을 추가해서 범례 지정

```python
# legend()를 사용하기 위해서 plot() 에서 label을 지정해야 호출이 가능하다.
plt.plot([1,2,3,4], [1,4,9,16], label='Price ($)')
plt.xlabel('X-label')
plt.ylabel('Y-label')
plt.legend()
plt.show()
```

## 위치 지정(loc)
- legend() 함수 내에 loc=(xpad, ypad) 또는 '위치'를 추가해서 위치 지정

```python
plt.plot([1,2,3,4], [1,4,9,16], label='Price ($)')
plt.xlabel('X-label')
plt.ylabel('Y-label')
# plt.legend(loc=(1, 1))
# plt.legend(loc=(0.5, 0.5))
# plt.legend(loc='lower right')
plt.legend(loc='upper left')
plt.show()
```

## 열 개수 지정
- ncol파라미터를 legend() 함수에 넣어서 열개수 지정

```python
# legend() 함수 내에서 ncol파라미터를 사용해 벌례 열 개수를 지정할 수 있다.
plt.plot([1,2,3,4], [1,4,9,16], label='Price($)')
plt.plot([1,2,3,4], [3,5,9,7], label='Demand(#)')
plt.xlabel('X-label')
plt.ylabel('Y-label')
plt.legend(loc='best', ncol=2)
plt.show()
```

## 범례 테두리 꾸미기
- legend() 함수 내에 frameon, shadow파라미터를 추가해서 사용
- frameon은 범례 텍스트 상자의 테두리를 표시할지의 여부를 지정한다.(default=True)
- shadow파라미터로 텍스트 상자에 그림자를 표시할 수 있다.(default=False)

```python
plt.plot([1,2,3,4], [1,4,9,16], label='Price($)')
plt.plot([1,2,3,4], [3,5,9,7], label='Demand(#)')
plt.xlabel('X-label')
plt.ylabel('Y-label')
plt.legend(loc='best', ncol=2, fontsize=10, frameon=True, shadow=True)
plt.show()
```
