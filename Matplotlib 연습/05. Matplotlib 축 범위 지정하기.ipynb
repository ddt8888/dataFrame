```python
import matplotlib.pyplot as plt
import numpy as np
```

## 축범위 지정
- matplotlib.pyplot 모듈의 xlim(), ylim(), axis() 함수를 사용하면 그래프의 X, Y축이 표시되는 범위를 지정가능
- xlim() - X축이 표시되는 범위를 지정하거나 반환
- ylim() - Y축이 표시되는 범위를 지정하거나 반환
- axis() - X, Y축이 표시되는 범위를 지정하거나 반환


## 기본 사용 - xlim(), ylim()

```python
# plt.xlim(xmin, xmax) / plt.xlim((xmin, xmax)) / plt.xlim([xmin, xmax]) 모두 가능
# plt.ylim(ymin, ymax) / plt.ylim((ymin, ymax)) / plt.ylim([ymin, ymax]) 모두 가능
plt.plot([1,2,3,4], [1,4,9,16])
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.xlim(0, 5)
plt.ylim(0, 20)
plt.show()
```

## 기본 사용 - axis()
- plt.axis((xmin, xmax, ymin, ymax))
- plt.axis([xmin, xmax, ymin, ymax])

```python
plt.plot([1,2,3,4], [1,4,9,16])
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.axis((0, 5, 0, 20))
plt.show()
```

## 옵션 지정하기
- xmin, xmax = plt.xlim()
- ymin, ymax = plt.ylim()
- xmin, xmax, ymin, ymax = plt.axis()

- axis() 옵션에는 아래의 다양한 옵션을 제공
    - 'on' | 'off' | 'equal' | 'scaled' | 'tight' | 'auto' | 'normal' | 'image' | 'square'

```python
plt.plot([1,2,3,4],[1,4,9,16])
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.axis('scaled')
# plt.axis('square')
# plt.axis('tight')
# plt.axis('auto')
plt.show()
```

## 축 범위 얻기

```python
plt.plot([1, 2, 3, 4], [2, 3, 5, 10])
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

x_range, y_range = plt.xlim(), plt.ylim()
print(x_range, y_range)

axis_range = plt.axis('scaled')
print(axis_range)

plt.show()
```
