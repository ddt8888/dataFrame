```python
import matplotlib.pyplot as plt
import numpy as np
```

#### 하나의 그래프 영역에 여러 개의 곡선을 나타내는 방법



## 기본 사용

```python
x = np.arange(0, 2, 0.2)

# plot(x1, y1, style1, x2, y2, style2, x3, y3, style3)
plt.plot(x, x, 'r--', x, x**2, 'bo', x, x**3, 'g-.')

plt.show()
```

```python
plt.plot(x, x, 'r--')
plt.plot(x, x**2, color='b', marker='^')
plt.plot(x, x**3, 'g-.')
```
