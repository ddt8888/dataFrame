```python
import matplotlib.pyplot as plt
import numpy as np
```

## 선종류
- Solid('-')
- Dashed('--')
- Dotted(':')
- Dash-dot('-.')


## 기본 사용

```python
plt.plot([1,2,3],[4,4,4], '-', label='Solid')
plt.plot([1,2,3],[3,3,3], '--', label='Dashed')
plt.plot([1,2,3],[2,2,2], ':', label='Dotted')
plt.plot([1,2,3],[1,1,1], '-.', label='Dash-dot')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.ylim(0,5)
plt.legend(ncol=4, loc='best')
plt.show()
```

## 기본 사용을 linestyle로 지정

```python
plt.plot([1,2,3],[4,4,4], linestyle='solid', label='Solid')
plt.plot([1,2,3],[3,3,3], linestyle='dashed', label='Dashed')
plt.plot([1,2,3],[2,2,2], linestyle='dotted', label='Dotted')
plt.plot([1,2,3],[1,1,1], linestyle='dashdot', label='Dash-dot')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.ylim(0,5)
plt.legend(ncol=4, loc='best')
plt.show()
```

## 선 끝모양 지정하기
- plot() 함수안에 solid_capstyle, dash_capstyle을 사용해 선의 끝모양 지정 가능

```python
plt.plot([1,2,3],[4,4,4], linestyle='solid', label='Solid+butt', linewidth=10, solid_capstyle='butt')
plt.plot([1,2,3],[3,3,3], linestyle='solid', label='Solid+round', linewidth=10, solid_capstyle='round')
plt.plot([1,2,3],[2,2,2], linestyle='dashed', label='Dashed+butt', linewidth=10, dash_capstyle='butt')
plt.plot([1,2,3],[1,1,1], linestyle='dashed', label='Dashed+round', linewidth=10, dash_capstyle='round')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.ylim(0,5)
plt.legend(ncol=2, loc='best')
plt.show()
```
