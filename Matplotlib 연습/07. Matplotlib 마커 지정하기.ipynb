```python
import matplotlib.pyplot as plt
import numpy as np
```

## Matplotlib 마커 지정하기
- plot() 함수의 포맷 문자열(Format String)을 사용해서 그래프의 선과 마커를 지정할 수 있음


## 기본 사용

```python
# 'bo'를 입력하면 파란색(b)의 원형(o) 마커로 표시가 된다.
plt.plot([1,2,3,4], [1,4,9,16], 'bo')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.axis((0, 5, 0, 20))
plt.show()
```

## 선과 마커 동시에 표시

```python
# plt.plot([1,2,3,4], [1,4,9,16], 'bo-')
# plt.plot([1,2,3,4], [1,4,9,16], 'bo--')
# plt.plot([1,2,3,4], [1,4,9,16], 'bo:')
plt.plot([1,2,3,4], [1,4,9,16], 'bo-.')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()
```

## Marker 파라미터 사용하기

```python
plt.plot([5,6,7], marker='H')
plt.plot([4,5,6], marker='d')
plt.plot([3,4,5], marker='x')
plt.plot([2,3,4], marker=11)
plt.plot([1,2,3], marker='$H$')
plt.show()
```
