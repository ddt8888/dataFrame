```python
import matplotlib.pyplot as plt
import numpy as np
```

matplotlib.pyplot 모듈의 xlabel(), ylabel() 함수를 사용하면 \
그래프 x, y축에 대한 레이블을 표시할수 있다.

## 기본 사용

```python
plt.plot([1,2,3,4], [1,4,9,16])
plt.xlabel('X-label')
plt.ylabel('Y-label')
plt.show()
```

## 여백 지정(labelpad)

```python
plt.plot([1,2,3,4], [1,4,9,16])
plt.xlabel('X-label', labelpad=15)
plt.ylabel('Y-label', labelpad=10)
plt.show()
```

## 폰트 설정(fontdict={...})

```python
plt.plot([1,2,3,4], [1,4,9,16])
plt.xlabel('X-label', labelpad=15, fontdict={'family': 'serif', 'color': 'b', 'weight': 'bold', 'size': 14})
plt.ylabel('Y-label', labelpad=10, fontdict={'family': 'fantasy', 'color': 'deeppink', 'weight': 'normal', 'size': 'xx-large'})
plt.show()
```

```python
# 위 폰트설정처럼 바로 적용할 수도 있지만
# 객체에 넣어서 사용하면 편리하게 재사용 할 수 있다.
font1 = {'family': 'serif', 'color': 'b', 'weight': 'bold', 'size': 14}
font2 = {'family': 'fantasy', 'color': 'deeppink', 'weight': 'normal', 'size': 'xx-large'}
plt.plot([1,2,3,4], [1,4,9,16])
plt.xlabel('X-label', labelpad=15, fontdict=font1)
plt.ylabel('Y-label', labelpad=10, fontdict=font2)
plt.show()
```

## 위치 지정하기(loc)

```python
plt.plot([1,2,3,4], [1,4,9,16])
# loc파라미터로 위치를 지정할 수 있다.
# xlabel() 함수에서는 loc는 'left', 'center', 'right'만 가능하고
# ylabel() 함수에서는 loc는 'top', 'center', 'bottom'이 가능하다.
plt.xlabel('X-label', labelpad=15, loc='right')
plt.ylabel('Y-label', labelpad=10, loc='top')
plt.show()
```
