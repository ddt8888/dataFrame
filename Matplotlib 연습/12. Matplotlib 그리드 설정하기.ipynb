```python
import matplotlib.pyplot as plt
import numpy as np
```

### 데이터의 위치를 더 명확하게 나타내기 위해 그래프에 그리드 (Grid, 격자)를 표시
#### grid() 함수 사용


## 기본사용

```python
x = np.arange(0, 2, 0.2)

plt.plot(x, x, 'bo')
plt.plot(x, x**2, color='#e35f62', marker='*')
plt.plot(x, x**3, color='forestgreen', marker='^')
# grid() 함수 내 True, False 입력 가능
# bool 값을 생략하면 default=True
plt.grid()
plt.show()
```

## 축 지정하기

- grid(True, axis='x')
- grid(True, axis='y')
- color로 색지정, alpha로 투명도, linestyle로 스타일 변경 가능

```python
x = np.arange(0, 2, 0.2)

plt.plot(x, x, 'bo')
plt.plot(x, x**2, color='#e35f62', marker='*')
plt.plot(x, x**3, color='forestgreen', marker='^')
# grid() 함수 내 True, False 입력 가능
# bool 값을 생략하면 default=True
# plt.grid(axis='y')
plt.grid(axis='x', color='red', alpha=0.5, linestyle='--')
plt.show()
```
