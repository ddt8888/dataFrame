```python
import matplotlib.pyplot as plt
import numpy as np
```

### 그래프의 특정 위치에 수직선/수평선을 표시하기 위해 matplotlib.pyplot 모듈은 아래의 네가지 함수를 지원

- axhline(): 축을 따라 수평선을 표시합니다.
- axvline(): 축을 따라 수직선을 표시합니다.
- hlines(): 지정한 점을 따라 수평선을 표시합니다.
- vlines(): 지정한 점을 따라 수직선을 표시합니다.



## 수평선 그리기 - axhline(), hlines()
- plt.axhline(y, xmin, xmax, colors, linestyle)
- plt.hline(y, xmin, xmax, colors, linestyle)

```python
x = np.arange(0, 4, 0.5)

plt.plot(x, x, 'bo')
plt.plot(x, x**2-4, 'g--')
plt.plot(x, -2*x+3, 'r:')
plt.axhline(4, 0.1, 0.9, color='gray', linestyle='--', alpha=0.5)
plt.hlines(-0.65, 1, 2.5, color='black', linestyle='-')
plt.show()
```

## 수직선 그리기 - axvline(), vlines()
- plt.axvline(x, ymin, ymax, color, linestyle)
- plt.vlines(x, ymin, ymax, color, linestyle)

```python
x = np.arange(0, 4, 0.5)

plt.plot(x, x, 'bo')
plt.plot(x, x**2-4, 'g--')
plt.plot(x, -2*x+3, 'r:')
plt.axvline(1, 0.1, 0.9, color='gray', linestyle='--', alpha=0.5)
plt.vlines(1.825, -3, 2, color='black', linestyle='-')
plt.show()
```
