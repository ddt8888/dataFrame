```python
import matplotlib.pyplot as plt
```

#### matplotlib.pyplot 모듈에서 그래프의 영역을 채우는 아래의 세가지 함수가 있다.
#### fill_between() / fill_betweenx() / fill()
- fill_between() - 두 수평 방향의 곡선 사이를 채웁니다.
- fill_betweenx() - 두 수직 방향의 곡선 사이를 채웁니다.
- fill() - 다각형 영역을 채웁니다.


## 기본사용 - fill_between()

```python
x = [1,2,3,4]
y = [1,4,9,16]
plt.plot(x, y)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
# fill_between() 함수에 x index 범위, y index 범위를 지정해서 영역을 채운다.
# 이때 채운 색깔이 너무 진하면 투명도를 나타내는 alpha를 넣어 투명도를 지정해준다.
plt.fill_between(x[1:3], y[1:3], alpha=0.5)
# 네 점 (x[1], y[1]), (x[2], y[2]), (x[1], 0), (x[2], 0)을 잇는 영역
plt.show()
```

## 기본사용 - fill_betweenx()

```python
x = [1,2,3,4]
y = [1,4,9,16]
plt.plot(x, y)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
# fill_between() 함수에 y index 범위, x index 범위를 지정해서 영역을 채운다.
# 이때 채운 색깔이 너무 진하면 투명도를 나타내는 alpha를 넣어 투명도를 지정해준다.
plt.fill_betweenx(y[2:4], x[2:4], alpha=0.5)
# 네 점 (x[2], y[2]), (x[3], y[3]), (0, y[2]), (0, y[3])을 잇는 영역
plt.show()
```

## 두 그래스 사이 영역 채우기

```python
x = [1, 2, 3, 4]
y1 = [2, 3, 5, 10]
y2 = [1, 2, 4, 8]

plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.fill_between(x[1:3], y1[1:3], y2[1:3], alpha=0.5)
plt.show()
```

```python
x = [1, 2, 3, 4]
y1 = [2, 3, 5, 10]
y2 = [1, 2, 4, 8]

plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.fill([1.9, 1.9, 3.1, 3.1], [1.0, 4.0, 6.0, 3.0], color='lightgray', alpha=0.5)

plt.show()
```
