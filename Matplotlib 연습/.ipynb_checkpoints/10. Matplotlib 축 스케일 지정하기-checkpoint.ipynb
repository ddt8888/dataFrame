```python
import matplotlib.pyplot as plt
import numpy as np
```

#### matplotlib.pyplot 모듈의 xscale(), yscale() 함수를 사용해서 그래프의 축 스케일을 지정 가능
#### scale() 함수 내에 파라미터는 기본적으로 linear로 표시되지만 log / symlog / logit으로 변경 가능



## X축 스케일 지정하기

```python
# np.linspace(start=배열의 시작값, stop=배열의 끝값, 
# num=start와 stop 사이를 몇 개의 일정한 간격으로 요소를 만들 것인지 나타내는 것)
# num 생략하면 default=50
x = np.linspace(-10, 10, 100)
y = x ** 3

plt.plot(x, y)
# symlog = Symmetrical Log Scale
# => 축의 원점을 기준으로 양, 음의 방향이 대칭적인 로그 스케일로 표시
plt.xscale('symlog')

plt.show()
print(x)
```

## Y축 스케일 지정하기

```python
x = np.linspace(0, 5, 100)
y = np.exp(x)

plt.plot(x, y)
plt.yscale('linear')
# plt.yscale('log')

plt.show()
```
