```python
import matplotlib.pyplot as plt
import numpy as np
```

### matplotlib.pyplot 모듈의 title() 함수 이용해 그래프의 타이틀(Title) 설정


## 기본 사용

```python
x = np.arange(0, 2, 0.2)

plt.plot(x, x, 'bo')
plt.plot(x, x**2, color='red', marker='*')
plt.plot(x, x**3, color='green', marker='^', markersize=9)
plt.xticks((0, 1, 2))
plt.yticks(np.arange(0, 7))
plt.title('Graph Title')
plt.show()
```

```python
x = np.arange(0, 2, 0.2)

plt.plot(x, x, 'bo')
plt.plot(x, x**2, color='red', marker='*')
plt.plot(x, x**3, color='green', marker='^', markersize=9)
plt.xticks((0, 1, 2))
plt.yticks(np.arange(0, 7))
# loc파라미터는 'left', 'right', 'center'가 있고 default는 center이다.
# pad파라미터는 그래프와 간격을 얼마나 두는지 지정
# font파라미터를 넣어 따로 폰트 지정 가능
plt.title('Graph Title', loc='right', pad=10)
plt.show()
```
