```python
import matplotlib.pyplot as plt
import numpy as np
```

### 파이 차트 (Pie chart, 원 그래프)는 범주별 구성 비율을 원형으로 표현한 그래프

## 기본 사용
- pie()

```python
ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']

# autopct는 부채꼴 안에 표시될 숫자의 형식을 지정
# %0.1f 는 소수점 1자리까지 표기, %%는 %문자를 그그대로 반환되게 함.
plt.pie(ratio, labels=labels, autopct='%0.1f%%')
plt.show()
```

## 시작 각도와 방향 설정하기
- startangle는 부채꼴이 그려지는 시작 각도를 설정(default=0(양의 방향x축))
- counterclock=False로 설정하면 시계 방향 순서로 부채꼴 영역이 표시

```python
ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']

# autopct는 부채꼴 안에 표시될 숫자의 형식을 지정
# %0.1f 는 소수점 1자리까지 표기, %%는 %문자를 그그대로 반환되게 함.
plt.pie(ratio, labels=labels, autopct='%0.1f%%', startangle=260, counterclock=False)
plt.show()
```

## 중심에서 벗어나는 정도 설정
- explode파라미터를 사용해서 벗어나게 할 수 있다.

```python
ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']
explode = [0, 0.1, 0, 0.1]

# autopct는 부채꼴 안에 표시될 숫자의 형식을 지정
# %0.1f 는 소수점 1자리까지 표기, %%는 %문자를 그그대로 반환되게 함.
plt.pie(ratio, labels=labels, autopct='%0.1f%%', startangle=260, \
        counterclock=False, explode=explode)
plt.show()
```

## 그림자 표시
- shadow파라미터를 사용해 그림자 효과 표시 가능(value는 True, False)

## 색상 표시
- colors파라미터를 사용해서 색 지정 가능

```python
ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']
explode = [0.05, 0.05, 0.05, 0.05]
colors=['silver', 'gold', 'whitesmoke', 'lightgray']

# autopct는 부채꼴 안에 표시될 숫자의 형식을 지정
# %0.1f 는 소수점 1자리까지 표기, %%는 %문자를 그그대로 반환되게 함.
plt.pie(ratio, labels=labels, autopct='%0.1f%%', startangle=260, \
        counterclock=False, explode=explode, shadow=True, colors=colors)
plt.show()
```

## 부채꼴 모양으로 그래프 생성
- wedgeprops 파라미터를 이용해서 부채꼴로 스타일 설정 가능
- wedgeprops{'width':value, 'edgecolor':'value', 'linewidth':value}

```python
ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']
colors=['silver', 'gold', 'whitesmoke', 'lightgray']
wedgeprops={'width':0.7, 'edgecolor':'w', 'linewidth':5}

# autopct는 부채꼴 안에 표시될 숫자의 형식을 지정
# %0.1f 는 소수점 1자리까지 표기, %%는 %문자를 그그대로 반환되게 함.
plt.pie(ratio, labels=labels, autopct='%0.1f%%', startangle=260, \
        counterclock=False, wedgeprops=wedgeprops)
plt.show()
```
