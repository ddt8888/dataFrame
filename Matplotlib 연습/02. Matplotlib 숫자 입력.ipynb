```python
import matplotlib.pyplot as plt
import numpy as np
```

- matplotlib.pyplot 모듈의 plot()은 선 또는 마커 그래프 그리기에 사용되는 함수
- plot()함수에 y값을 리스트 형태로 입력하면 위 그림과 같은 꺾은선 그래프가 그려짐

## 기본 사용

```python
# 아래와 같이 하나의 리스트 형태로 값을 입력하면 y값으로 인식한다.
# plt.plot((2,3,5,10))(<= 파이썬 튜플) 또는 
# plt.plot(np.array([2,3,5,10]))(<= Numpy array)로도 입력가능
plt.plot([2,3,5,10])
plt.show()
```

## x, y값 입력하기

```python
# 아래 2개의 리스트를 입력하면 순서대로 x, y로 인식해서
# (1,1)(2,4)(3,9)(4,16)를 잇는 꺾은선 그래프가 나타난다.
plt.plot([1,2,3,4],[1,4,9,16])
plt.show()
```

## 레이블이 있는 데이터 사용

```python
# 딕셔너리 생성
data_dic={'data_x': [1,2,3,4,5], 'data_y': [2,3,5,10,8]}
# x,y 축에 데이터의 레이블(딕셔너리의 키)을 입력하고 
# data 파라미터에 딕셔너리 지정
plt.plot('data_x', 'data_y', data=data_dic)
plt.show()
```
