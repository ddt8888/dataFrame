```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
df
```

#### Hist Plot

```python
df['total_bill'].values
```

```python
sns.histplot(x=df['total_bill'].values)
# == sns.displot(x=df['total_bill'], kind='hist')  # 구버전
```

```python
sns.histplot(x=df['total_bill'], y=df['tip'])
# == sns.displot(x=df['total_bill'], y=df['tip'], kind='hist')  # 구버전
```

#### KDE Plot
- kdeplot : 밀도 추정치 (histplot : 절대량) / 곡선

```python
sns.kdeplot(x=df['total_bill'])
# = sns.displot(x=df['total_bill'], kind='kde')
```

```python
sns.kdeplot(x=df['total_bill'], y=df['tip'])
# = sns.displot(x=df['total_bill'], y=df['tip'], kind='kde')
```

#### ECDF Plot - 누적 분포

```python
sns.ecdfplot(x=df['total_bill'])
# == sns.displot(x=df['total_bill'], kind='ecdf')
```

#### Rug Plot

```python
sns.kdeplot(x=df["total_bill"])
sns.rugplot(x=df["total_bill"])
```

#### Count Plot

```python
sns.countplot(x = df['sex'])
```

#### Box Plot

```python
sns.boxplot(x = df['total_bill'])
```

```python
sns.boxplot(y = df['total_bill'], x = df['smoker'])
```

#### Violin Plot

```python
sns.violinplot(y = df['total_bill'], x = df['smoker'])
```

#### Strip Plot

```python
sns.stripplot(y = df['total_bill'], x = df['smoker'])
```

#### Swarm Plot

```python
sns.swarmplot(y = df['total_bill'], x = df['smoker'])
```

#### Heat Map

```python
df.corr()
```

```python
sns.heatmap(df.corr(), annot = True, cmap = 'ocean_r')
```

#### Joint Plot

```python
# scatter(산점도) / hist
sns.jointplot(x=df['total_bill'], y=df['tip'], kind = 'scatter')
```

#### Pair Plot

```python
sns.pairplot(df)
```

```python
# hue 적용
sns.pairplot(df, hue='smoker', palette="Set1")
```

#### Reg Plot

```python
sns.regplot(x = 'tip', y = 'total_bill', data = df)
```

#### LM Plot
- regplot()과 faceGrid를 결합

```python
sns.lmplot(x = 'tip', y = 'total_bill', data = df, hue = 'sex')
```
