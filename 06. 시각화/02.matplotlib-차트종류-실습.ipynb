### 막대 그래프 : https://wikidocs.net/92095

```python
x = np.arange(3)
print(x)
years = ['2018', '2019', '2020']  # X축에 표시될 연도
values = [100, 400, 900]          # 막대 그래프의 y 값

plt.bar(x, values)    # x:x축 값, values:y축 값
plt.xticks(x, years)  # years는 label

plt.show()
```

### 산점도 : https://wikidocs.net/92110
- 산점도 (Scatter plot)는 두 변수의 상관 관계를 직교 좌표계의 평면에 점으로 표현하는 그래프입니다.

```python
np.random.seed(0)

n = 50
x = np.random.rand(n)   # [0, 1) 범위의 난수 50개
y = np.random.rand(n)
# print(x)
# print(y)

plt.scatter(x, y)
plt.show()
```

```python
# 색상과 크기
np.random.seed(0)

n = 50
x = np.random.rand(n)
y = np.random.rand(n)
area = (30 * np.random.rand(n))**2
colors = np.random.rand(n)
# print(area)
 
# s는 마커의 크기 : 마커의 크기는 size**2 의 형태로 지정 => s=20**2는 markersize=20과 동일
# c는 마커의 color : 데이터의 길이와 같은 크기의 숫자 시퀀스 또는 rgb, 그리고 Hex code 색상을 입력해서 지정
plt.scatter(x, y, s=area, c=colors) 
plt.show()
```

```python
# markersize=20과 크기 비교 => s=20**2 과 동일
plt.plot([1], [1], 'o', markersize=20, c='#FF5733')
plt.scatter([2], [1], s=20**2, c='#33FFCE')

plt.text(0.5, 1.05, 'plot(markersize=20)', fontdict={'size': 14})
plt.text(1.6, 1.05, 'scatter(s=20**2)', fontdict={'size': 14})
plt.axis([0.4, 2.6, 0.8, 1.2])
plt.show()
```

```python
# 투명도와 컬러맵
np.random.seed(0)

n = 50
x = np.random.rand(n)
y = np.random.rand(n)
area = (30 * np.random.rand(n))**2
colors = np.random.rand(n)

# alpha : 마커의 투명도 / 0에서 1 사이의 값
# cmap : 컬러맵에 해당하는 문자열
# Spectral가 아닌 이상한 값을 넣으면 입력 가능한 값이 표시 됨
plt.scatter(x, y, s=area, c=colors, alpha=0.5, cmap='YlOrBr')  
plt.colorbar()
plt.show()
```

### 파이 차트 : https://wikidocs.net/92114

```python
ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']

plt.pie(ratio, labels=labels, autopct='%.1f%%')  # autopct : 숫자의 형식 / 소수점 1자리
plt.show()
```

```python
# 시작 각도와 방향
ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']
 
# startangle : 시작 각도 / 디폴트는 0도 (양의 방향 x축)로 설정
# counterclock=False : 시계 방향 순서로 부채꼴 영역이 표시
plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False)
plt.show()
```

```python
# 중심에서 벗어나는 정도
ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']
explode = [0, 0.10, 0, 0]

# explode : 중심에서 벗어나는 정도
plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False, explode=explode)
plt.show()
```

```python
# 그림자
ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']
# explode = [0.05, 0.05, 0.05, 0.05]
explode = [0, 0.10, 0, 0]

# shadow=True : 그림자
plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False, explode=explode, shadow=True) 
plt.show()
```

```python
# 색상
ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']
# explode = [0.05, 0.05, 0.05, 0.05]
explode = [0, 0.10, 0, 0]
colors = ['silver', 'gold', 'whitesmoke', 'lightgray']

plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False, explode=explode, shadow=True, colors=colors)
plt.show()
```

```python
ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']
# explode = [0.05, 0.05, 0.05, 0.05]
explode = [0, 0.10, 0, 0]
colors = ['#ff9999', '#ffc000', '#8fd9b6', '#d395d0']

plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False, explode=explode, shadow=True, colors=colors)
plt.show()
```

```python
# 부채꼴 스타일
ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']
colors = ['#ff9999', '#ffc000', '#8fd9b6', '#d395d0']
# 부채꼴 영역의 너비 (반지름에 대한 비율), 테두리의 색상, 테두리 선의 너비
wedgeprops={'width': 0.7, 'edgecolor': 'w', 'linewidth': 5}

plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False, colors=colors, wedgeprops=wedgeprops)
plt.show()
```

### 여러 개의 그래프 그리기 : https://wikidocs.net/141537

```python
x1 = np.linspace(0.0, 5.0)  # (start, stop, num) - num이 생략되면 50개
x2 = np.linspace(0.0, 2.0)

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)  # np.cos - 삼각함수  / np.pi - 3.14.... / np.exp - 지수함수
y2 = np.cos(2 * np.pi * x2)

# (행, 렬, 현재순서)
plt.subplot(2, 1, 1)                # nrows=2, ncols=1, index=1
plt.plot(x1, y1, 'o-')
plt.title('1st Graph')
plt.ylabel('Damped oscillation')

plt.subplot(2, 1, 2)                # nrows=2, ncols=1, index=2
plt.plot(x2, y2, '.-')
plt.title('2nd Graph')
plt.xlabel('time (s)')
plt.ylabel('Undamped')

plt.tight_layout()
plt.show()
```

```python
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

plt.subplot(1, 2, 1)                # nrows=1, ncols=2, index=1
plt.plot(x1, y1, 'o-')
plt.title('1st Graph')
plt.xlabel('time (s)')
plt.ylabel('Damped oscillation')

plt.subplot(1, 2, 2)                # nrows=1, ncols=2, index=2
plt.plot(x2, y2, '.-')
plt.title('2nd Graph')
plt.xlabel('time (s)')
plt.ylabel('Undamped')

plt.tight_layout()
plt.show()
```

### 텍스트 삽입하기 : https://wikidocs.net/141543

```python
a = 2.0 * np.random.randn(10000) + 1.0
b = np.random.standard_normal(10000)
c = 20.0 * np.random.rand(5000) - 10.0

plt.hist(a, bins=100, density=True, alpha=0.7, histtype='step')  # density=True => 빈도수가 아닌 밀도
plt.text(1.0, 0.35, '2.0*np.random.randn(10000)+1.0')  # (x위치, y위치, text)
plt.hist(b, bins=50, density=True, alpha=0.5, histtype='stepfilled')
plt.text(2.0, 0.20, 'np.random.standard_normal(10000)')
plt.hist(c, bins=100, density=True, alpha=0.9, histtype='step')
plt.text(5.0, 0.08, 'np.random.rand(5000)-10.0')
plt.show()
```

```python
# 행과 열 설정
fig, ax = plt.subplots(2, 2)  # 행, 렬
plt.show()
```

```python
# 그래프 그리기
x = np.arange(1, 5)     # [1, 2, 3, 4]

fig, ax = plt.subplots(2, 2)
ax[0][0].plot(x, np.sqrt(x))      # left-top
# ax[0][1].plot(x, x)               # right-top
ax[1][0].plot(x, -x+5)            # left-bottom
ax[1][1].plot(x, np.sqrt(-x+5))   # right-bottom

plt.show()
```

```python
# 스타일 설정
x = np.arange(1, 5)     # [1, 2, 3, 4]

fig, ax = plt.subplots(2, 2, sharex=True, sharey=True, squeeze=True)
ax[0][0].plot(x, np.sqrt(x), 'gray', linewidth=3)
ax[0][1].plot(x, x, 'g^-', markersize=10)
ax[1][0].plot(x, -x+5, 'ro--')
ax[1][1].plot(x, np.sqrt(-x+5), 'b.-.')

plt.show()
```

```python
# 제목과 범례
x = np.arange(1, 5)     # [1, 2, 3, 4]

fig, ax = plt.subplots(2, 2, sharex=True, sharey=True, squeeze=True)
ax[0][0].plot(x, np.sqrt(x), 'gray', linewidth=3, label='y=np.sqrt(x)')
ax[0][0].set_title('Graph 1')
ax[0][0].legend()
ax[0][1].plot(x, x, 'g^-', markersize=10, label='y=x')
ax[0][1].set_title('Graph 2')
ax[0][1].legend(loc='upper left')
ax[1][0].plot(x, -x+5, 'ro--', label='y=-x+5')
ax[1][0].set_title('Graph 3')
ax[1][0].legend(loc='lower left')
ax[1][1].plot(x, np.sqrt(-x+5), 'b.-.', label='y=np.sqrt(-x+5)')
ax[1][1].set_title('Graph 4')
ax[1][1].legend(loc='upper center')

plt.show()
```
