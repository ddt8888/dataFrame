#### dropna() - 결측치가 있는 행을 모두 삭제
- 참조자료 : https://cosmosproject.tistory.com/308

- DataFrame.dropna(axis=0/1, how='any'/'all', subset=[col1, col2, ...], inplace=True/False)

- axis = 0/1
    - 0 : NaN 값이 포함된 row를 drop (default 값입니다.)
    - 1 : NaN 값이 포함된 column을 drop

- how = 'any'/'all'
    - any : row 또는 column에 NaN값이 1개만 있어도 drop (default)
    - all : row 또는 column에 있는 모든 값이 NaN이어야 drop

- inplace = True/False

- subset = [col1, col2, ...]
    - subset을 명시하지 않으면 DataFrame 전체(모든 column & 모든 row)에 대해 dropna를 진행
    - subset을 명시하면 subset에 적힌 column값에 대해서만 dropna를 진행

```python
# row에 nan값이 하나라도 있으면 해당 row 삭제
df.dropna()
```

```python
df_1 = df.dropna()
# df_1.info()
df_1.isna().sum()
```

```python
df.dropna(subset=['Age', 'Cabin'])
```

```python
df.dropna(subset=['Age', 'Cabin']).isna().sum()
```

```python
# row의 모든 값이 nan일때 해당 row를 삭제
df.dropna(how='all')
```

```python
df.dropna(how='any')
```

```python
df.shape
```

```python
# row값이 하나라도 nan일때 해당 row를 삭제
df.dropna(axis=0, how='any', subset=['Age', 'Cabin'], inplace=True)
df
```

- 특정값 대체
- 평균값으로 대체
- 앞방향, 뒷방향 값으로 대체
- inplace=True

- 특정값 대체

```python
df['Age'].fillna(0)
```

```python
df['Age'].fillna(0).isna().sum()
```

```python
# 명시된 컬럼에 특정 값
# dic_val = {'Age':0, 'Cabin':'no-val'}
# df.fillna(dic_val)
df.fillna({'Age':0, 'Cabin':'no-val'})
```

```python
df.fillna({'Age':0, 'Cabin':'no-val'}).isna().sum()
```

- 평균값으로 대체

```python
mean_age = df['Age'].mean()
mean_age
```

```python
df['Age'].fillna(mean_age)
```

```python
df['Age'].fillna(df['Age'].mean())
```

- 사망자는 사망자 평균나이로 생존자는 생존자 평균나이로 fillna

```python
# 생존자/사망자의 평균나이
# df['Survived'] == 1
# df[df['Survived'] == 1]
# df[df['Survived'] == 1]['Age']
mean_age_s1 = df[df['Survived'] == 1]['Age'].mean()
mean_age_s0 = df[df['Survived'] == 0]['Age'].mean()
print(f'생존자의 평균나이 : {mean_age_s1}')
print(f'사망자의 평균나이 : {mean_age_s0}')
```

```python
# 생존자/사망자의 평균나이로 fill
# 생존자
# df[df['Survived'] == 1]['Age']
# df[df['Survived'] == 1]['Age'].isna().sum()
# df[df['Survived'] == 1]['Age'].fillna(mean_age_s1)
# df[df['Survived'] == 1]['Age'].fillna(mean_age_s1).isna().sum()
df[df['Survived'] == 1]['Age'].fillna(mean_age_s1)
```

```python
# df['Survived'] == 1
# df.loc[df['Survived'] == 1, 'Age']
df.loc[df['Survived'] == 1, 'Age'] = df[df['Survived'] == 1]['Age'].fillna(mean_age_s1)
df.loc[df['Survived'] == 0, 'Age'] = df[df['Survived'] == 0]['Age'].fillna(mean_age_s0)
```

```python
df['Age'].isna().sum()
```

#### 앞방향, 뒷방향 값으로 대체
- 한눈에 보기 위해 다른 Data로 테스트
- method='pad'   : 결측치 바로 이전의 값
- method='bfill' : 결측치 바로 뒤의 값
- limit : 대체하는 개수

```python
df2 = pd.DataFrame({
    'c1' : [10, np.nan, np.nan, 30],
    'c2' : [40, 50, 60, np.nan],
    'c3' : [np.nan, 70, 80, 90],
    }
)
df2
```

```python
# 결측치 바로 이전의 값으로
df2.fillna(method='pad')
```

```python
# 결측치 바로 뒤의 값으로
df2.fillna(method='bfill')
```

```python
# limit = 1 => 1개만 대체
df2.fillna(method='pad', limit=1)
```
