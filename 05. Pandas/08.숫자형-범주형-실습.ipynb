### 숫자형, 범주형
#### 숫자형 : 연속성을 가지는 숫자
- Age, Fare

#### 범주형 : 연속적이지 않은 값
- Name, Sex, Cabin, 등
- 숫자형태의 값이지만 범주형에 속하는 것
    - Survived, Pclass, 등

#### 숫자형 -> 범주형

- astype
    - Pclass를 범주형으로(int64 -> object)

```python
df = df.astype({'Pclass':'object'})
df.info()
```

```python
df = df.astype({'Pclass':str})
df.info()
```

```python
df['Pclass'] = df['Pclass'].astype(str)
```

#### pd.to_numeric
- Pclass를 숫자형으로

```python
df['Pclass'] = pd.to_numeric(df['Pclass'])
df.info()
```

```python
df['Pclass'] = df['Pclass'].astype(str)
```

```python
df.loc[0, 'Pclass'] = 'a'
df.head()
```

```python
df['Pclass'] = pd.to_numeric(df['Pclass'])
```

```python
df['Pclass'] = pd.to_numeric(df['Pclass'], errors='ignore')
df.info()
```

```python
# errors='coerce' 오류 발생하지 않고 NaN값으로 변경하여 처리
df['Pclass'] = pd.to_numeric(df['Pclass'], errors='coerce')
df.info()
```

### apply함수 : 전체의 원소에 원하는 연산

- Age를 세대별로

```python
import math
```

```python
# 나이를 세대별로
def age_cate(age):
    if math.isnan(age):
        return '나이미상'
    else:
        firstage = str(int(age/10))
        if firstage == '0':
            return '10대미만'
        else:
            return firstage + '0대'
```

```python
df['연령대'] = df['Age'].apply(age_cate)
df.head()
```

```python
df[df['Age'].isna()]['연령대'].value_counts()
```

#### lambda 함수

```python
# 일반 함수
def hap(x, y):
    return x + y

hap(10, 20)
```

```python
# lambda
# lambda x, y: x + y   => x, y가 매개변수, x + y 가 함수 기능
(lambda x, y: x + y)(10, 20)
```

#### 의미없는 값이지만 lambda와 apply 연습

```python
# if 한줄 ==> 결과 = A if 조건 else B
df['연령여부'] = df['Age'].apply(lambda x : 'N' if math.isnan(x) else 'Y')
df.head(10)
```

```python
df['연령대3'] = df['Age'].apply(lambda x : '나이미상' if math.isnan(x) else ( '10대미만' if int(x/10) == 0  else ( str(int(x/10)) + '0대')))
df.head(10)
```
