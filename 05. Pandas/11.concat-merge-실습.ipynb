```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('./train.csv')
df.head()
```

### Concat - Concatenating(연결)

- sample data 만들기

```python
df1 = df.loc[:9, ['PassengerId', 'Survived', 'Pclass', 'Name']]
df1
```

```python
df2 = df.loc[10:19, ['PassengerId', 'Survived', 'Pclass', 'Name']]
df2
```

#### column명이 같을 경우

```python
# column명이 같을 경우
pd.concat([df1, df2])
```

#### column명이 다를 경우

```python
# column명을 다르게
df3 = df2.copy()
df3.columns = ['PassengerId', 'Survived', 'Pclass', 'Name2']
df3
```

```python
# column명이 다를 경우
pd.concat([df1, df3])
```

#### axis = 1 => 동일한 index를 기준으로 가로로 연결

```python
pd.concat([df1, df2], axis=1)
```

```python
df4 = df2.reset_index()
df4
```

```python
df5 = df2.copy()
df5.index = [0, 1, 2, 3, 4, 5, 6, 10, 11, 12]
df5
```

```python
df1
```

```python
pd.concat([df1, df4], axis=1)
```

```python
pd.concat([df1, df5], axis=1)
```

### merge - sql의 join과 비슷한 개념
- on : join 시 비교할 컬럼
- how : join 방법 -> inner, left, right, outer (sql join과 동일 개념)

- sample data 만들기

```python
df2.loc[:15, 'PassengerId'] = np.arange(1,7)
df2
```

```python
df2.columns = ['PassengerId', 'c1', 'c2', 'c3']
df2
```

```python
# inner
pd.merge(df1, df2, on='PassengerId', how='inner')
```

```python
df1
```

```python
pd.merge(df1, df2, on='PassengerId', how='left')
```

```python
pd.merge(df1, df2, on='PassengerId', how='right')
```

```python
pd.merge(df1, df2, on='PassengerId', how='outer')
```

- 2개 이상의 column, column명이 다를 때

```python
df1.head(2)
```

```python
df2.head(2)
```

```python
pd.merge(df1, df2, left_on=['PassengerId', 'Survived'], right_on=['PassengerId', 'c1'], how='inner')
```
